```python
import matplotlib.pyplot as plt

marks = {'Andy':88, 'Amy':66, 'James':90, 'Jules':55, 'Arthur':77}

plt.bar(marks.keys(), marks.values())
plt.title('Student Grades')
plt.xlabel('Students')
plt.ylabel('Grades')
plt.show()

plt.pie(marks.values(), labels=marks.keys(), autopct='%1.1f%%')
plt.title('Student Grades')
plt.legend(title='Students', loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4])
y1 = np.array([0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4])

x2 = np.array([0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4])
y2 = np.array([1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5])

x3 = np.array([0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4])
y3 = np.array([2, 2.5, 3, 3.5, 4, 4.5, 5, 5.5, 6])

plt.subplot(2, 3, 1)
plt.plot(x1, y1)
plt.plot(x2, y2, color = 'red')
plt.plot(x3, y3, color = 'blue')
plt.title('Multple Lines')

x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([10, 11, 12, 13, 14, 15])

plt.subplot(2,3,2)
plt.bar(x,y)
plt.title('Bar')

x = np.array([1, 2, 3, 4, 5, 6])
plt.subplot(2, 3, 3)
plt.pie(x)
plt.title('Pie')

x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([1, 2, 3, 4, 5, 6])
plt.subplot(2, 3, 4)
plt.plot(x,y)
plt.grid()
plt.title('Grid')

x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([10, 11, 12, 13, 14, 15])

plt.subplot(2, 3, 5)
plt.scatter(x,y)
plt.title('Scatter')

x = np.array([1, 2, 3, 4, 5, 6])
plt.subplot(2, 3, 6)
plt.hist(x)
plt.title('Histogram')
```
